```python
def segment_stability_across_solutions(data, min_clusters, max_clusters):
    segment_memberships = {}
    
    for n_clusters in range(min_clusters, max_clusters + 1):
        kmeans = KMeans(n_clusters=n_clusters, random_state=1234)
        labels = kmeans.fit_predict(data)
        segment_memberships[n_clusters] = labels
    
    stability_scores = {}
    
    for n_clusters in range(min_clusters + 1, max_clusters + 1):
        prev_labels = segment_memberships[n_clusters - 1]
        curr_labels = segment_memberships[n_clusters]
        stability_scores[n_clusters] = adjusted_rand_score(prev_labels, curr_labels)
    
    return stability_scores


segment_stability_scores = segment_stability_across_solutions(MD_x, min_clusters, max_clusters)
```

```python

stability_df = pd.DataFrame(list(segment_stability_scores.items()), columns=['Number of Segments', 'Adjusted Rand Index'])


plt.figure(figsize=(10, 6))
sns.lineplot(data=stability_df, x='Number of Segments', y='Adjusted Rand Index', marker='o')
plt.xlabel('Number of Segments')
plt.ylabel('Segment Stability (Adjusted Rand Index)')
plt.title('Segment Stability Across Solutions')
plt.grid(True)
plt.show()
```

```python
def segment_stability_within_solution(data, n_clusters, nrep=10):
    kmeans = KMeans(n_clusters=n_clusters, random_state=1234)
    original_labels = kmeans.fit_predict(data)
    
    segment_stability = []
    
    for cluster in range(n_clusters):
        stability_scores = []
        for _ in range(nrep):
            kmeans = KMeans(n_clusters=n_clusters, random_state=np.random.randint(10000))
            labels = kmeans.fit_predict(data)
            cluster_labels = (labels == cluster)
            original_cluster_labels = (original_labels == cluster)
            stability_scores.append(adjusted_rand_score(original_cluster_labels, cluster_labels))
        segment_stability.append(np.mean(stability_scores))
    
    return segment_stability


n_clusters = 4
segment_stability = segment_stability_within_solution(MD_x, n_clusters)
```

```python

plt.figure(figsize=(10, 6))
plt.bar(range(1, n_clusters + 1), segment_stability)
plt.xlabel('Segment Number')
plt.ylabel('Segment Stability')
plt.title('Segment Level Stability Within Solutions')
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

```python


min_clusters = 2
max_clusters = 8
nrep = 10 
```

```python
def fit_mixture_model(data, n_clusters, nrep=10):
    best_model = None
    best_score = -np.inf
    for _ in range(nrep):
        gmm = GaussianMixture(n_components=n_clusters, random_state=np.random.randint(10000), covariance_type='diag')
        gmm.fit(data)
        score = gmm.score(data)
        if score > best_score:
            best_model = gmm
            best_score = score
    return best_model
```

```python
models = {}
criteria = []

for n_clusters in range(min_clusters, max_clusters + 1):
    model = fit_mixture_model(MD_x, n_clusters, nrep)
    models[n_clusters] = model
    log_lik = model.score(MD_x) * len(MD_x)  # Log likelihood
    aic = model.aic(MD_x)
    bic = model.bic(MD_x)
    icl = bic  # ICL is often approximated as BIC in simpler implementations
    criteria.append((n_clusters, log_lik, aic, bic, icl))

criteria_df = pd.DataFrame(criteria, columns=['n_clusters', 'log_likelihood', 'AIC', 'BIC', 'ICL'])
```

```python
plt.figure(figsize=(12, 8))
plt.plot(criteria_df['n_clusters'], criteria_df['AIC'], marker='o', label='AIC')
plt.plot(criteria_df['n_clusters'], criteria_df['BIC'], marker='o', label='BIC')
plt.plot(criteria_df['n_clusters'], criteria_df['ICL'], marker='o', label='ICL')
plt.xlabel('Number of Components')
plt.ylabel('Value of Information Criteria (AIC, BIC, ICL)')
plt.legend()
plt.title('Information Criteria for Mixture Models')
plt.grid(True)
plt.show()
```

```python
selected_n_clusters = 4
mixture_model = models[selected_n_clusters]
mixture_clusters = mixture_model.predict(MD_x)


np.random.seed(1234)
kmeans_clusters = np.random.randint(4, size=len(MD_x))


cross_tab = pd.crosstab(kmeans_clusters, mixture_clusters)
print(cross_tab)
```

```python

initial_labels = kmeans_clusters
mixture_model_initialized = GaussianMixture(n_components=selected_n_clusters, random_state=1234, covariance_type='diag')
mixture_model_initialized.fit(MD_x, initial_labels)

mixture_clusters_initialized = mixture_model_initialized.predict(MD_x)


cross_tab_init = pd.crosstab(kmeans_clusters, mixture_clusters_initialized)
print(cross_tab_init)
```

```python
log_lik_rand_init = mixture_model.score(MD_x) * len(MD_x)
log_lik_kmeans_init = mixture_model_initialized.score(MD_x) * len(MD_x)

print(f'Log Likelihood (Random Initialization): {log_lik_rand_init}')
print(f'Log Likelihood (k-means Initialization): {log_lik_kmeans_init}')
```

```python

np.random.seed(1234)
mcdonalds = pd.DataFrame(np.random.randint(0, 2, size=(1453, 11)), columns=['yummy', 'convenient', 'spicy', 'fattening', 'greasy',
                                                                           'fast', 'cheap', 'tasty', 'expensive', 'healthy', 'disgusting'])

like_mapping = {'I hate it!-5': -5, '-4': -4, '-3': -3, '-2': -2, '-1': -1, '0': 0, '+1': 1, '+2': 2, '+3': 3, '+4': 4, 'I love it!+5': 5}
mcdonalds['Like'] = np.random.choice(list(like_mapping.keys()), size=len(mcdonalds))
mcdonalds['Like_n'] = mcdonalds['Like'].apply(lambda x: like_mapping[x])
```

```python

formula = 'Like_n ~ ' + ' + '.join(mcdonalds.columns[:-2])

gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=1234)
gmm.fit(mcdonalds.iloc[:, :-2])

clusters = gmm.predict(mcdonalds.iloc[:, :-2])


mcdonalds['Cluster'] = clusters
```

```python

models = []
for cluster in range(2):
    cluster_data = mcdonalds[mcdonalds['Cluster'] == cluster]
    y = cluster_data['Like_n']
    X = cluster_data.iloc[:, :-3]  
    X = sm.add_constant(X) 
    model = sm.OLS(y, X).fit()
    models.append(model)

for idx, model in enumerate(models):
    print(f"Component {idx + 1}:")
    print(model.summary())
```

```python

plt.figure(figsize=(10, 6))
for idx, model in enumerate(models):
    coef = model.params.drop('const') 
    err = np.abs(model.conf_int().iloc[1:, :] - coef.values.reshape(-1, 1)) 

    
    err = err.T 

    
    plt.errorbar(coef.index, coef.values, yerr=err, fmt='o', label=f'Component {idx + 1}', capsize=5)

plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Perception Variables')
plt.ylabel('Regression Coefficients')
plt.title('Regression Coefficients of Two-Segment Mixture of Linear Regression Models')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
